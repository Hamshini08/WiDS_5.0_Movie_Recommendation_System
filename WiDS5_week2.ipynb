```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

movies = pd.read_csv("C:/Users/WIN/OneDrive/Desktop/WiDS 5.0/tmdb_5000_movies.csv")
movies.head()
```

```python
def extract_genres(genre_str):
    genre_names = [] #empty list
    parts = genre_str.split('"name": "')
    for part in parts[1:]:
        genre = part.split('"')[0]
        genre_names.append(genre)
    return genre_names

movies['genre_list'] = movies['genres'].apply(extract_genres) #stores all genres in a coloumn
all_genres = [] 
for genres in movies['genre_list']:
    all_genres.extend(genres) #list of all genres

genre_counts = pd.Series(all_genres).value_counts(sort=False) #convert the list to series and count the frequency of each genre

plt.figure(figsize=(14,6))
genre_counts.plot(kind='bar')
plt.title("Number of movies per genre")
plt.xlabel("Genre")
plt.ylabel("Number of movies")
plt.tight_layout()
plt.show()
 
```
