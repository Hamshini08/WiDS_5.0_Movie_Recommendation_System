```python
import pandas as pd
df1 = pd.read_csv(r"C:\Users\WIN\OneDrive\Desktop\WiDS 5.0\movies_metadata.csv",low_memory = False)
df2 = pd.read_csv(r"C:\Users\WIN\OneDrive\Desktop\WiDS 5.0\links.csv")
df3 = pd.read_csv(r"C:\Users\WIN\OneDrive\Desktop\WiDS 5.0\keywords.csv")
df4 = pd.read_csv(r"C:\Users\WIN\OneDrive\Desktop\WiDS 5.0\credits.csv")

df1 = df1.dropna(subset=['id'])
df1['id'] = df1['id'].astype(str) 

df2 = df2.dropna(subset=['tmdbId'])
df2['tmdbId'] = df2['tmdbId'].astype(int).astype(str) #converting tmdbid to a string
df2 = df2.rename(columns={'tmdbId':'id'}) 

df3 = df3.dropna(subset=['id'])
df3['id'] = df3['id'].astype(str)

df4 = df4.dropna(subset=['id'])
df4['id'] = df4['id'].astype(str)

merged_df = df1.merge(df3, on='id', how='inner') \
               .merge(df4, on='id', how='inner') 
final_df = merged_df[merged_df['id'].isin(df2['id'])]
print(final_df.columns)
print(final_df.shape)
```
